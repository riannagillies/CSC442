4. use pandas **describe** function to caculate descriptie statistics of all quantitative columns,
Find out what functions can be used for calculating spread

```python
# import libraries
import pandas as pd

# Use describe function to calculate descriptive stats of quantitative columns
df_penguin.describe()

# Which functions can be used to calculate spread
print("Variances: ")
print(df_penguin.describe().var())

print("Ranges: ")
print(df_penguin.describe().max() - df_penguin.describe().min())
```

8. For all categorical columns (textual data) caculate **value_counts** which shows frequency of each category in the column

```python
# Select all categorical columns
categorical_cols = df_penguin.select_dtypes(include='object').columns

# Get value counts for each categorical column
for col in categorical_cols:
    print(f"Value counts for {col}:")
    print(df_penguin[col].value_counts())
    print("\n")
```
